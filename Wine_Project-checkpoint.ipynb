# Look for common characteristics of wine choices

```python
## Define a function to find wines that are exactly like current choices

def rec_basic(df):
    
    """
    Look for the wines that match the most common characteristics of the ones that we know
    that Arielle already enjoys.
    
    This includes setting up year, price, variety, and country filters. For each unique variety, we'll 
    find a high and low priced wine within our range that is the top rated and fits our other
    filtered variables.
    
    The function will return a data frame with two recommendations per wine.
    """
    
    current_wine = df[df['arielle_choice'] == 1]
    
    current_filter = (df['arielle_choice'] != 1) # We don't want a wine that she already has tried
    
    year_min = current_wine.year.min()
    year_max = current_wine.year.max()
    
    year_filter = ((df.year >= year_min) & (df.year <= year_max)) # Filter for the year range of wines that she enjoys
    
    price_min = current_wine.price.min()
    price_max = current_wine.price.max()
    
    price_mid = price_min + ((price_max - price_min) / 2) # Create midpoint for high/low price adjuster
    
    price_filter = ((df.price >= price_min) & (df.price <= price_max)) # Filter for her typical price points
    
    countries = list(set(current_wine.country))
    
    country_filter = (df.country.isin(countries)) # Filter for country
    
    varieties = list(set(current_wine.variety))
    
    variety_filter = df.variety.isin(varieties) # Filter for wine type
        
    filtered_df = df[current_filter & year_filter & price_filter & country_filter & variety_filter] # Filtered data frame
    
    recommendation = []
    
    for variety in varieties: # Recommendation for each of her favorite varieties 
        
        var_df = filtered_df[filtered_df.variety == variety]
        
        lower_price = var_df[var_df.price <= price_mid] # One pick that's below the mid-point pricing
        higher_price = var_df[var_df.price > price_mid] # and one that's above
        
        top_rec_low = lower_price[['country','designation','points','price','title','variety']].sort_values(
            by='points',ascending=False)[:1] # Extract the highest rated lower-priced wine
        top_rec_high = higher_price[['country','designation','points','price','title','variety']].sort_values(
            by='points',ascending=False)[:1] # Extract the highest rated higher price wine
        
        recommendation.extend(top_rec_low.index) # Add the index value of the lower priced wine to the list
        recommendation.extend(top_rec_high.index) # Add the index value of the higher priced wine to the list
        
    rec_df = df.loc[df.index.isin(recommendation),:] # Extract only the recommendation index values from the data
    
    rec_df.sort_values(by=['variety', 'price'], ascending=True, inplace=True) # Sort/group by wine type then price
    
    return rec_df # Return the recommendation data frame
```

# Recommendation with All Text

```python
def recommender_with_text(df):
    
    """
    Takes all of the columns for a single observations, groups the values into a single string, then compares
    each of the unique values/words in the string to a string of text from all of the combined observations
    that collectively comprise our recommendation variables.
    
    The wines are then scaled and ranked based on the number of similar words that they have to the 
    """
    
    df_wine_text = df

    listed_cols = [col for col in df_wine_text.columns if col != 'arielle_choice'] # We can't include her choices in our recommendation
    df_wine_text.dropna(subset=['price'],axis=0,inplace=True) # Price column is not helpful here
    
    listed_rows = df_wine_text[listed_cols].values.tolist()
    
    all_text_row = [''.join(str(i)) for i in listed_rows] # Create a single string with the values
    
    # Remove unnecessary filler characters
    
    all_text_row = [i.replace(',','').replace('\\','').replace('[','').replace(']','') for i in all_text_row]
    
    # Add a column to the data frame with all of the text
    df_wine_text['all_text'] = all_text_row
    
    ## Natural language Processing dictionaries that we'll use

    from nltk.tokenize import word_tokenize, regexp_tokenize
    import nltk
    from nltk.corpus import stopwords
    
    # Assign the stop words to a variable
    stop_words = set(stopwords.words('english'))
    
    all_air_text = df_wine_text[df_wine_text['arielle_choice'] == 1] # Extract all columns that were her choices
    
    air_text = ''.join(string for string in all_air_text['all_text']) # Join the text in all of her choices into one string
    
    air_text = air_text.replace(r"'",'').replace(r'(','').replace(r')','') # Replace unnecessary characters
    
    pattern = r'\w+|\d+' # Write a pattern for matching both text or digits 

    regex = regexp_tokenize(air_text, pattern) # Apply pattern to air_text variable
    
    unique_regex = list(set(regex)) # List of unique words in the full combined description
    unique_regex = [i for i in unique_regex if i not in stop_words] # Remove stop words from the unique regex
    
    num_same = []

    for i in df_wine_text['all_text']: # Iterate over the values in all_text and count the number of words that also appear
                                       # in our string of unique words for Arielle's preferred wines
    
        counter = 0
    
        text = ''.join(string for string in i) # Join all text together into a string
        regex_2 = regexp_tokenize(text, pattern) # Apply tokenization to the list of strings
    
        for word in list(set(regex_2)): # If the word is in the full description of Arielle text, increase the counter by 1
        
            if word in unique_regex:
                counter +=1
            
            else:
                continue
            
        num_same.append(counter)
        
    # Plot the distribution of unique values
    
    plt.hist(num_same, bins=20)
    plt.title('Number of matching words per wine')
    plt.ylabel('Number of occurences')
    plt.xlabel('Number of words')
    plt.show()

    print('The maximum number of matching words in a description is: %d words' %max(num_same))
    
    df_wine_text['similarity'] = num_same # Add the similarity column back to our original data frame
    
    ## Divide the number of words by the length of total potential matching words to turn the words into a ratio

    df_wine_text['similarity'] = df_wine_text.similarity.divide(len(unique_regex))
    
    ## Filter out her choices so that we're only looking at new options

    new_choices = df_wine_text[df_wine_text['arielle_choice'] == 0]
    
    ## Sort the new values by the similarity metric, then by points

    new_sorted = new_choices.sort_values(by=['similarity','points'],ascending=False)
    
    ## We still want to make sure we stay with the same variety, so we create a list of those

    preferred_variety = list(set(df_wine[df_wine['arielle_choice'] == 1].variety))
    
    ## Filter the data set for her preferred varieties

    new_sorted_var = new_sorted[new_sorted.variety.isin(preferred_variety)]
    
    # Import a scaler method and apply it to the data
    
    from sklearn.preprocessing import MinMaxScaler

    sclr = MinMaxScaler()
    new_sorted_var['similarity'] = list(sclr.fit_transform(np.asarray(new_sorted_var.similarity).reshape(-1,1)))
    
    ## Format the newly scaled values

    new_sorted_var['similarity'] = [round(float(i),3) for i in new_sorted_var.similarity]
    
    ## Re-sort by our scaled values

    new_sorted_var = new_sorted_var.sort_values(by=['similarity','points'],ascending=False)
    
    ## Only include relevant columns

    cols_to_keep = ['country','designation','points','price','province','region_1','region_2','title','variety','winery','year','similarity']

    new_sorted_var = new_sorted_var[cols_to_keep]
    
    # Find the top two wines for each variety and append it to a data frame
    
    final_recs = pd.DataFrame()
    
    for var in preferred_variety:
        
        variety_df = new_sorted_var[new_sorted_var.variety == var]
        final_recs = pd.concat([final_recs, variety_df[:2]], axis=0)
    
    return final_recs
```

# Recommendation 3: Description only

We're still biasing our results toward the U.S. because by including all the text, we're including region and country, which skew toward the U.S. in our sample data. Therefore, we'll use the review, which might help offset some of this.

Rather than largely building from scratch, we'll utilize sklearn's TFIDF Vectorizer feature to convert our text data to numeric variables that we can use to judge similarity between a given wine and the others in the data set. For a full explanation, please refer to the accompanying article.

# TFIDF Vectorizer

```python
## Import TFIDF libraries

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## Condense the file for the sake of being able to run the analysis locally and because we still want to only apply
## the analysis to the known varieties that Arielle likes.

paired_file = df_wine[df_wine.variety.isin(list(set(df_wine[df_wine.arielle_choice == 1]['variety'])))]
```

```python
def tfidf_recommendation(df):
    
    """
    Takes in a data frame with wine descriptions, passes the descriptions into a TFIDF function, determines
    the similarity of a given observation/description to the rest of the inputs using cosine similarity, 
    and return a data frame with the top similarity scores for each of our wines.
    """
    
    # Extract the description column from the entered data
    
    all_descriptions = df[['description']]
    
    # Initialize a TFIDF Vectorizer model to work with the text data
    
    tf = TfidfVectorizer(analyzer='word',
                     min_df=0,
                     stop_words='english')

    # Use the initiated TFIDF model to transform the data in descriptions
    
    tfidf_matrix = tf.fit_transform(all_descriptions['description'])
    
    # Compute the cosine similarities between the items in the newly transformed TFIDF matrix
    cosine_similarities = cosine_similarity(tfidf_matrix,tfidf_matrix)
    
    # Reset the index of the data frame to be able to iterate through the index with our smaller data frame
    
    arielle_indeces = df.reset_index()
    arielle_indeces = arielle_indeces[arielle_indeces.arielle_choice == 1]

    # Initialize a dictionary to store the results
    
    results = {} 

    # Iterate through the observation of Arielle's choices
    
    for idx, row in arielle_indeces.iterrows():

        similar_indices = cosine_similarities[idx].argsort()[:-5:-1] # Extract the top 5 wines for each observation that Arielle likes
        similar_items = [(cosine_similarities[idx][i], df.reset_index()['index'][i]) for i in similar_indices] # Find the TFIDF score of that item
        results[row['index']] = similar_items[1:] # Append all results after the first (which will be itself) to the dictionary
        
    tfidf_recs = pd.DataFrame()

    # Iterate through the dictionary of results to add the recommended values to a data frame
    
    for k, v in results.items():
    
        for i in v:
        
            tfidf_recs = tfidf_recs.append(df[df.index.isin([i[1]])])
            
    tfidf_scores = []

    # Iterate through the data frame of recommended wines to find their TFIDF scores and add that to the data frame
    
    for i in tfidf_recs.index:
    
        for k, v in results.items():
        
            for ele in v:
            
                if i == ele[1]:
            
                    tfidf_scores.append(ele[0])
                
    tfidf_recs['tfidf_score'] = tfidf_scores
    
    return tfidf_recs
```

```python
# Apply the new TFIDF predictor to our paired file

tfidf_recs = tfidf_recommendation(paired_file)
```

```python
## Drop nan prices and filter for below $50

final_recs = final_recs.dropna(subset=['price'], axis=0)
final_recs = final_recs[final_recs.price <= 50]
```

```python
## I know she likes GSM, and I like Syrah, so we'll limit to those varities

varities_final = ['GSM','Syrah']

final_recs = final_recs[final_recs.variety.isin(varities_final)]
```

```python
## Rating above 90

final_recs = final_recs[final_recs.points >= 90]
```

```python
## For our final recommendation, I want one GSM and one Syrah, with the goal of one being local and one international. 
# After examining the filtered data, there is only one international wine left, and it's a GSM from Australia. 
## This means that our Syrah choice will be the local one. There are only two Syrah's local - one in California 
## and one in Washington. We'll use the one with a higher points rating as our recommendation.

final_recs = final_recs[(final_recs.country != 'US') | (final_recs.points == final_recs[(final_recs.country == 'US') & 
                                                                                        (final_recs.variety == 'Syrah')].points.max())]
```
